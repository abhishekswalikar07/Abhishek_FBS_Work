## 11. Find the transpose of matrix A and matrix B.

```python
A=np.arange(1,5).reshape(2,2)
print(A)
B=np.arange(5,9).reshape(2,2)
print(B)
print("Transpose of A:\n", A.T)
print("Transpose of B:\n", B.T)
```
