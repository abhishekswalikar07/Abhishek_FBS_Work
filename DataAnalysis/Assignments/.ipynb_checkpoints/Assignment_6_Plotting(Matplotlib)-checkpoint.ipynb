```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Dataset: The Cars93 dataset contains information about different car models, including their manufacturer, price, horsepower, MPG, and other attributes. You will use this dataset for the following tasks.

## 1. Bar Chart: 
## a. Create a bar chart to compare the average price of cars by manufacturer. 
## b. Label the axes and add a suitable title. 
## c. Use different colors for different manufacturers.

```python
data=pd.read_csv('../Datasets/cars93.csv')
```

```python
data
```

```python
x=data.groupby('Manufacturer')['Price'].mean().index
y=data.groupby('Manufacturer')['Price'].mean().values
```

```python
plt.figure()
plt.bar(x,y)
plt.xlabel("Manufacturer")
plt.ylabel("Average Price")
plt.title("Average Car Price by Manufacturer")
plt.xticks(rotation=90)
plt.show()
```

### 2. Line Chart:
### a. Create a line chart showing the trend of horsepower for different types of cars.
### b. Include proper legends, labels, and a title

```python
data.head()
```

```python
x=data.groupby('Type')['Horsepower'].mean().index
y=data.groupby('Type')['Horsepower'].mean().values
```

```python
plt.figure()
plt.plot(x,y,marker='o')
plt.xlabel("Car Type")
plt.ylabel("Average Horsepower")
plt.title("Horsepower Trend by Car Type")
plt.legend(["Horsepower"])
plt.show()
```

### 3. Pie Chart:
### a. Create a pie chart showing the proportion of different car types in the dataset.
### b. Include percentages in the chart.

```python
data.head()
```

```python
x=data['Type'].value_counts()
y=data['Type'].value_counts().index
```

```python
plt.pie(x, labels=y, autopct='%0.2f%%')
plt.title("Proportion of Car Types")
plt.show()
```

### 4. Scatter Plot:
### a. Create a scatter plot to show the relationship between horsepower and MPG (Miles Per Gallon).
### b. Label the axes and add a title.

```python
plt.figure()
plt.scatter(x='Horsepower',y='MPG.city',data=data)
plt.xlabel("Horsepower")
plt.ylabel("MPG (City)")
plt.title("Horsepower vs MPG")
plt.show()
```
