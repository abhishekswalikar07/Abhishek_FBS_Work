# 1.Refer to googleplaystore.csv file and solve following questions

## a. Display First 5 records from Dataframe.


```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('E:/FBS/DataAnalysis/Datasets/googleplaystore.csv')
data
```

## b. Display category with highest number of apps registered.


```python
# data['Category'].max()
category_counts = data['Category'].value_counts()
print("\nCategory with highest number of apps:")
print(category_counts.idxmax(), "-", category_counts.max())
```

## c. What is the average rating for all the apps


```python
avg_rating = np.mean(data['Rating'].dropna())
print("\nAverage rating of all apps:")
print(avg_rating)
```

## d. Display total number of apps which even supports Andriod version 2.3 and up


```python
android_23 = data[data['Android Ver'].str.contains("2.3", na=False)]
print("\nTotal apps supporting Android 2.3 and up:")
print(android_23.shape[0])
```

## e. Impute the missing values available in rating column with respective measure of central tendency.

```python
missing_before = data['Rating'].isnull().sum()
data['Rating'] = data['Rating'].fillna(avg_rating)
missing_after = data['Rating'].isnull().sum()

print("\nMissing values in Rating column:")
print("Before imputation:", missing_before)
print("After imputation:", missing_after)
```
