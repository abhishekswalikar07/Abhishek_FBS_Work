```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('../../Datasets/Position_Salaries.csv')
```

```python
data.head()
```

```python
X=data[['Level']]
y=data['Salary']
```

## Check with linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
pred=lr.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score(y,pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Level'],data['Salary'])
plt.plot(data['Level'],pred)
```

## Implementing polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=4)
```

```python
X_trans=poly.fit_transform(X)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_trans,y)
```

```python
poly_predict=lr.predict(X_trans)
```

```python
r2_score(y,poly_predict)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data['Level'],data['Salary'])
plt.plot(data['Level'],poly_predict,marker='p',markerfacecolor='red')
# plt.plot(data['Level'],poly_predict)
plt.show()
```

```python
level=6
# values=[[1,6,36,216,1296]]
values=poly.transform([[level]])
lr.predict(values)
```
