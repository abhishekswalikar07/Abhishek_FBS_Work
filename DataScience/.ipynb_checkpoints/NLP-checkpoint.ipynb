## Stemming/ Lemmatization 

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
words=["eating", "eats", "ate", "adjustable", "rafting", "ability", "meeting","better","beutiful"]
```

```python
for word in words:
    print(word,"  ",ps.stem(word))
```

```python
text='I am meeting him tomorrow at the meeting'
words=text.split(" ")
words
```

```python
for word in words:
    print(word,' ',ps.stem(word))
```

```python
text
```

```python
doc=nlp(text)
for token in doc:
    print(token,' ',token.lemma_,' ',token.pos_)
```

```python
words
```

```python
text=' '.join(words)
text
```

```python
doc=nlp(text)
for token in doc:
    print(token,' ',token.lemma_)
```

```python
# text = "Tesla Inc has acquired Twitter Inc in India"
text = "Mr. Smith is a wise person"
doc=nlp(text)
for token in doc.ents:
    print(token,' ',token.label_)
```

## Stop word removal

```python

stop_words=nlp.Defaults.stop_words
# print(stop_words)
```

```python
len(stop_words)
```

```python
stop_words.remove('do')
```

```python
len(stop_words)
```

```python
nlp.vocab['do'].is_stop=False
```

```python
# 'offer' in stop_words
nlp.vocab['offer'].is_stop
```

```python
stop_words.add('offer')
```

```python
len(stop_words)
```

```python
nlp.vocab['offer'].is_stop=True
```

```python
text
```

```python
doc=nlp(text)
words=[]
for token in doc:
    if(token.lemma_ not in stop_words):
        words.append(token.lemma_)
new_text=' '.join(words)
```
