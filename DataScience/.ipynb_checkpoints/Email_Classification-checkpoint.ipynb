```python
import pandas as pd
import spacy
```

```python
data=pd.read_csv('../DataScience/Datasets/spam.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data['Category'].value_counts()
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
stop_words=nlp.Defaults.stop_words
len(stop_words)
stop_words.remove('do')
nlp.vocab['do'].is_stop=False
```

```python
corpus=[]
for i in range(0,data.shape[0]):
    words=[]
    doc=nlp(data['Message'][i])
    for token in doc:
        if(token.lemma_ not in stop_words):
            words.append(token.lemma_)
    new_text=' '.join(words)
    corpus.append(new_text)
```

```python
new_text
```

```python
data['Message'][0]
```

```python
# corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
output=tf.fit_transform(corpus)
output
```

```python
output=output.toarray()
```

```python
# unique words from corpu which will columns in input matrix
# tf.get_feature_names_out()
# We are printing  50 words
print(tf.get_feature_names_out()[2000:2050])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(output,data['Category'],test_size=0.3,random_state=45)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
model=MultinomialNB()
model.fit(X_train,y_train)
```

```python
print('Training acc:',accuracy_score(model.predict(X_train),y_train))
print('Testing acc:',accuracy_score(model.predict(X_test),y_test))
```
