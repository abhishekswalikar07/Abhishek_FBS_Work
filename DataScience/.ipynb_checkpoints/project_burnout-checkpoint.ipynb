## check outliers

```python
def checkOutliers(data, col):
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.boxplot(data[col])
    plt.title(f'boxplot of {col}')

    plt.subplot(1,2,2)
    sns.histplot(data=data, x=col, kde=True)
    
    mean = data[col].mean()
    std = data[col].std()

    plt.axvline(mean + 3*std, color='red')
    plt.axvline(mean + 3*std, color='red')

    plt.tight_layout()
    plt.show()
```

## Handling outliers

```python
def handleOutliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    data.loc[data[col] > upper, col] = upper
    data.loc[data[col] < lower, col] = lower
```

```python
handleOutliers(data, 'meetings_count')
```

```python
checkOutliers(data, 'meetings_count')
```

```python
handleOutliers(data, 'sleep_hours')
```

```python
checkOutliers(data, 'sleep_hours')
```

## checking VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculatVIF(data):
    vif = []
    output = pd.DataFrame()
    output['Attripute'] = data.columns
    for i in range(0, data.shape[1]):
        vif.append(variance_inflation_factor(data, i))
    output['VIF'] = vif
    return output
```

```python
X = data.drop(['user_id','burnout_score', 'burnout_risk'], axis=1)
y = data['burnout_risk']
```

```python
calculatVIF(X)
```

## data balancing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```python
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print("Before SMOTE:")
print(y_train.value_counts())

print("After SMOTE:")
print(y_train_sm.value_counts())
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(random_state=42)
```

```python
param_grid_rf = { "n_estimators": [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200],
                  "criterion": ["gini", "entropy"]}
```

```python
grid_search_rf = GridSearchCV( estimator=rf, param_grid=param_grid_rf, n_jobs=-1)
```

```python
grid_search_rf.fit(X_train_sm, y_train_sm)
```

```python
print("Best n_estimators:", grid_search_rf.best_params_)
```

```python
from sklearn.model_selection import cross_validate
```

```python
rf = RandomForestClassifier(criterion='gini', n_estimators=300)
result = cross_validate(rf, X_train_sm, y_train_sm)
rf.fit(X_train_sm, y_train_sm)
```

```python
result['test_score'].mean()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
test_pred = rf.predict(X_test)
accuracy_score(y_test, test_pred)
```

```python
confusion_matrix(y_test, test_pred)
```

```python
print(classification_report(y_test,test_pred))
```
