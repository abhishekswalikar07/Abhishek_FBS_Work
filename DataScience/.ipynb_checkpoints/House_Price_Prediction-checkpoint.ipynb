```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculate(data):
    vif=[]
    output=pd.DataFrame()
    output['Attribute']=data.columns
    for i in range(0,data.shape[1]):
        vif.append(variance_inflation_factor(data,i))
    output['vif']=vif
    return output
```

```python
calculate(data)
```
