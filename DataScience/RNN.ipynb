```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical
```

```python
text='Hello, we are training RNN model for character prediction'
text=text.lower()

chars=sorted(list(set(text)))
vocab_size=len(chars)

char_ind={c:i for i,c in enumerate(chars)}
ind_chars={i:c for i,c in enumerate(chars)}

print('Unique Char Size:',vocab_size)
print(ind_chars)
```

```python
from ast import Num
seq_length=5

X,y=[],[]

for i in range(len(text)-seq_length):
    seq=text[i:i+seq_length]
    X.append([char_ind[c] for c in seq])
    y.append(char_ind[text[i+seq_length]])

X=to_categorical(X,num_classes=vocab_size)
y=to_categorical(y,num_classes=vocab_size)

print('Training sample',len(X))
X.shape
```

```python
model=Sequential()
model.add(SimpleRNN(128,input_shape=(seq_length,vocab_size)))
model.add(Dense(vocab_size,activation='softmax'))
model.compile('adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X,y,epochs=50)
```

```python
def generate_text(input_char,length=1):
    text_out=input_char
    for i in range(length):
        seq=[char_ind[c] for c in input_char]
        seq=to_categorical(seq,num_classes=vocab_size)
        pred=model.predict(np.expand_dims(seq,axis=0))
        pred_char=ind_chars[np.argmax(pred)]
        text_out+=pred_char
        input_char=input_char[1:]+pred_char
    return text_out
input_char='train'
print(generate_text(input_char,length=5))
```
